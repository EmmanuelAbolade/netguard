## Section 1 — Theory: K-Means Clustering

### 1.1 What is K-Means?

K-Means is an **unsupervised** clustering algorithm. Unlike SVM and KNN which learn from labelled data, K-Means receives no class information — it discovers structure entirely from the data itself.

From the course notes: *'Unsupervised: no information is provided to the algorithm on which data points belong to which clusters.'*

Key distinction:
- **K-Nearest Neighbours** — supervised classification (uses labels)
- **K-Means Clustering** — unsupervised clustering (no labels)

### 1.2 The Algorithm

From the lecture notes:

1. **Initialise:** Pick k random data points as initial centroids
2. **Assign:** Place each point in the cluster whose centroid it is nearest to
3. **Update:** Recompute each centroid as the mean of all assigned points
4. **Repeat** steps 2-3 until convergence — points stop moving between clusters

*'The k-means algorithm will always reach a fixed point — a state in which clusters do not change'*

### 1.3 The Objective Function

K-Means minimises the **Within-Cluster Sum of Squares (WCSS)** — also called inertia:

WCSS = sum over all clusters k of: sum over all points x in cluster k of: ||x - mu_k||^2

Where mu_k is the centroid of cluster k. This is the sum of squared distances from each point to its assigned centroid. The algorithm iteratively reduces this until convergence.

**Important:** K-Means finds only a **local minimum** — the final result depends on initialisation. As stated in lectures: *'k-means converges, however only a local minimum is obtained.'*

### 1.4 Sensitivity to Initialisation

Poor initial centroid placement can lead to suboptimal clusters. Solutions from the lecture notes:

- **Multiple runs** — run K-Means several times with different random initialisations
- **K-Means++** — smarter initialisation that spreads centroids apart
- **Hierarchical clustering** — use as initialisation for K-Means

Scikit-learn uses **K-Means++** by default (`init='k-means++'`).

### 1.5 Choosing K — Elbow Method

From the lecture notes: *'The Elbow Method is a graphical representation of finding the optimal k in k-means clustering. It works by finding WCSS (Within-Cluster Sum of Square) i.e. the sum of the square distance between points in a cluster and the cluster centroid.'*

Plot WCSS vs k — the 'elbow' point where improvement slows dramatically suggests the optimal k.

### 1.6 Silhouette Method

A more rigorous measure of cluster quality. For each point i:

1. **a(i)** = average distance to all other points in the same cluster
2. **b(i)** = average distance to all points in the nearest other cluster
3. **s(i)** = (b(i) - a(i)) / max(a(i), b(i))

| Silhouette score | Meaning |
|---|---|
| Close to +1 | Well separated from neighbouring clusters |
| Close to 0 | On the boundary between clusters |
| Negative | Possibly assigned to wrong cluster |

### 1.7 Strengths and Weaknesses

| Strengths | Weaknesses |
|---|---|
| Low complexity O(nkt) | Must specify k in advance |
| Easy to implement | Sensitive to outliers |
| Scales to large datasets | Sensitive to initial centroids |
| Interpretable clusters | Only finds spherical clusters |
| Works well with numerical data | Not deterministic |

## Section 3 — From-Scratch K-Means Implementation

K-Means built using only NumPy — implementing the algorithm steps directly from the lecture notes.

```python
class KMeansScratch:
    """
    K-Means Clustering implemented from scratch using NumPy.

    Implements the algorithm from the lecture notes:
    1. Initialise k centroids randomly from the data
    2. Assign each point to its nearest centroid
    3. Recompute centroids as mean of assigned points
    4. Repeat until convergence (centroids stop moving)

    Minimises the WCSS objective:
    WCSS = sum_k sum_{x in C_k} ||x - mu_k||^2
    """

    def __init__(self, k=5, max_iters=100, tol=1e-4, random_state=42):
        self.k            = k
        self.max_iters    = max_iters
        self.tol          = tol
        self.random_state = random_state
        self.centroids    = None
        self.labels_      = None
        self.wcss_history = []

    def _euclidean(self, a, b):
        """Euclidean distance from point a to each row of matrix b."""
        return np.sqrt(((a - b) ** 2).sum(axis=1))

    def _assign(self, X):
        """Assign each point to its nearest centroid."""
        distances = np.array([
            self._euclidean(c, X) for c in self.centroids
        ]).T  # shape: (n_samples, k)
        return np.argmin(distances, axis=1)

    def _wcss(self, X, labels):
        """Compute Within-Cluster Sum of Squares (objective function)."""
        total = 0.0
        for k in range(self.k):
            mask = labels == k
            if mask.sum() > 0:
                diff   = X[mask] - self.centroids[k]
                total += (diff ** 2).sum()
        return total

    def fit(self, X):
        """Fit K-Means to data X using iterative centroid updates."""
        rng = np.random.RandomState(self.random_state)

        # Step 1: Initialise centroids by randomly sampling k data points
        idx = rng.choice(len(X), self.k, replace=False)
        self.centroids = X[idx].copy()

        for iteration in range(self.max_iters):
            # Step 2: Assign points to nearest centroid
            labels = self._assign(X)

            # Record WCSS
            wcss = self._wcss(X, labels)
            self.wcss_history.append(wcss)

            # Step 3: Update centroids as mean of assigned points
            new_centroids = np.array([
                X[labels == k].mean(axis=0) if (labels == k).sum() > 0
                else self.centroids[k]
                for k in range(self.k)
            ])

            # Step 4: Check convergence — centroids stopped moving
            shift = np.sqrt(((new_centroids - self.centroids) ** 2).sum())
            self.centroids = new_centroids

            if shift < self.tol:
                print(f"Converged after {iteration + 1} iterations.")
                break

        self.labels_ = self._assign(X)
        return self

    def predict(self, X):
        """Assign new points to nearest centroid."""
        return self._assign(X)


# Run from-scratch K-Means with k=5 (matching true number of classes)
print("Running from-scratch K-Means (k=5)...")
scratch_km = KMeansScratch(k=5, max_iters=100, random_state=42)
scratch_km.fit(X_scaled)

wcss_final = scratch_km.wcss_history[-1]
print(f"Final WCSS (objective): {wcss_final:,.2f}")
print(f"Cluster sizes: {dict(zip(*np.unique(scratch_km.labels_, return_counts=True)))}")
```

```python
# Visualise from-scratch convergence and clusters
fig, axes = plt.subplots(1, 2, figsize=(13, 4))

# WCSS convergence
axes[0].plot(scratch_km.wcss_history, "o-", color="#1E88E5", lw=2, ms=5)
axes[0].set_xlabel("Iteration")
axes[0].set_ylabel("WCSS (Within-Cluster Sum of Squares)")
axes[0].set_title("From-Scratch K-Means — WCSS Convergence")

# Discovered clusters in PCA space
for cluster_id in range(5):
    mask = scratch_km.labels_ == cluster_id
    axes[1].scatter(X_pca[mask,0], X_pca[mask,1],
                    c=palette[cluster_id % len(palette)],
                    s=8, alpha=0.5, label=f"Cluster {cluster_id}")

# Plot centroids in PCA space
centroids_pca = pca.transform(scratch_km.centroids)
axes[1].scatter(centroids_pca[:,0], centroids_pca[:,1],
                c="black", s=150, marker="X", zorder=5, label="Centroids")
axes[1].set_xlabel("PC1")
axes[1].set_ylabel("PC2")
axes[1].set_title("From-Scratch K-Means — Discovered Clusters")
axes[1].legend(fontsize=8, markerscale=2)

plt.suptitle("K-Means From Scratch — NSL-KDD Network Traffic", fontsize=13, y=1.02)
plt.tight_layout()
plt.savefig("images/kmeans_scratch.png", dpi=150, bbox_inches="tight")
plt.show()
```

## Section 4 — Scikit-learn Professional Implementation

```python
# ─────────────────────────────────────────────────────────────────────────────
# Elbow Method — find optimal k
# ─────────────────────────────────────────────────────────────────────────────

k_range   = range(2, 12)
wcss_vals = []
sil_vals  = []

print("Computing WCSS and Silhouette scores for k=2 to 11...")
for k in k_range:
    km = KMeans(n_clusters=k, init="k-means++", n_init=10,
                random_state=42, max_iter=300)
    km.fit(X_scaled)
    wcss_vals.append(km.inertia_)
    sil_vals.append(silhouette_score(X_scaled, km.labels_,
                                      sample_size=2000, random_state=42))
    print(f"  k={k:2d}: WCSS={km.inertia_:>12,.1f}  Silhouette={sil_vals[-1]:.4f}")
```

```python
# ─────────────────────────────────────────────────────────────────────────────
# Visualise Elbow Method and Silhouette Scores
# ─────────────────────────────────────────────────────────────────────────────

fig, axes = plt.subplots(1, 2, figsize=(13, 4))

# Elbow plot
axes[0].plot(list(k_range), wcss_vals, "o-", color="#1E88E5", lw=2, ms=6)
axes[0].set_xlabel("Number of Clusters (k)")
axes[0].set_ylabel("WCSS (Inertia)")
axes[0].set_title("Elbow Method — Finding Optimal k")
axes[0].set_xticks(list(k_range))

# Silhouette plot
best_k_sil = list(k_range)[np.argmax(sil_vals)]
axes[1].plot(list(k_range), sil_vals, "o-", color="#43A047", lw=2, ms=6)
axes[1].axvline(best_k_sil, color="gray", ls="--", alpha=0.7,
                label=f"Best k={best_k_sil}")
axes[1].set_xlabel("Number of Clusters (k)")
axes[1].set_ylabel("Silhouette Score")
axes[1].set_title("Silhouette Score — Higher is Better")
axes[1].set_xticks(list(k_range))
axes[1].legend()

plt.suptitle("Optimal k Selection — K-Means on NSL-KDD", fontsize=13, y=1.02)
plt.tight_layout()
plt.savefig("images/kmeans_elbow_silhouette.png", dpi=150, bbox_inches="tight")
plt.show()

print(f"Best k by Silhouette: {best_k_sil}")
print(f"Best Silhouette Score: {max(sil_vals):.4f}")
```

```python
# ─────────────────────────────────────────────────────────────────────────────
# Fit final K-Means model with k=5 (matching known classes)
# and evaluate cluster quality
# ─────────────────────────────────────────────────────────────────────────────

final_km = KMeans(n_clusters=5, init="k-means++", n_init=20,
                  random_state=42, max_iter=300)
final_km.fit(X_scaled)
cluster_labels = final_km.labels_

print("Final K-Means Model (k=5, k-means++ init, 20 restarts):")
print(f"  WCSS (inertia)    : {final_km.inertia_:,.2f}")
print(f"  Silhouette score  : {silhouette_score(X_scaled, cluster_labels):.4f}")
print(f"  Adjusted Rand Index: {adjusted_rand_score(y_true, cluster_labels):.4f}")
print(f"  Iterations to converge: {final_km.n_iter_}")

print("\nCluster sizes:")
for cid, cnt in zip(*np.unique(cluster_labels, return_counts=True)):
    print(f"  Cluster {cid}: {cnt:>5} samples ({cnt/len(cluster_labels):.1%})")
```

```python
# ─────────────────────────────────────────────────────────────────────────────
# Visualisation — Cluster comparison with true labels
# ─────────────────────────────────────────────────────────────────────────────

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# True labels
for i, (name, col) in enumerate(zip(class_names, palette)):
    mask = y_true == i
    axes[0].scatter(X_pca[mask,0], X_pca[mask,1],
                    c=col, s=6, alpha=0.4, label=name)
axes[0].set_title("True Labels (ground truth)")
axes[0].set_xlabel("PC1"); axes[0].set_ylabel("PC2")
axes[0].legend(markerscale=2, fontsize=9)

# K-Means discovered clusters
for cid, col in enumerate(palette[:5]):
    mask = cluster_labels == cid
    axes[1].scatter(X_pca[mask,0], X_pca[mask,1],
                    c=col, s=6, alpha=0.4, label=f"Cluster {cid}")

# Cluster centres in PCA space
centres_pca = pca.transform(final_km.cluster_centers_)
axes[1].scatter(centres_pca[:,0], centres_pca[:,1],
                c="black", s=200, marker="X", zorder=5, label="Centres")
axes[1].set_title("K-Means Discovered Clusters\n(No labels used during training)")
axes[1].set_xlabel("PC1"); axes[1].set_ylabel("PC2")
axes[1].legend(markerscale=2, fontsize=9)

plt.suptitle("K-Means Clustering vs Ground Truth — NSL-KDD", fontsize=13, y=1.02)
plt.tight_layout()
plt.savefig("images/kmeans_clusters_vs_truth.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
# ─────────────────────────────────────────────────────────────────────────────
# Silhouette plot — per-sample silhouette coefficients
# ─────────────────────────────────────────────────────────────────────────────

# Sample for tractable computation
n_sil = min(3000, len(X_scaled))
idx_sil = np.random.choice(len(X_scaled), n_sil, replace=False)
X_sil = X_scaled[idx_sil]
labels_sil = cluster_labels[idx_sil]

sil_vals_per_sample = silhouette_samples(X_sil, labels_sil)

fig, ax = plt.subplots(figsize=(8, 6))
y_lower = 10

for i in range(5):
    ith_sil = np.sort(sil_vals_per_sample[labels_sil == i])
    size     = ith_sil.shape[0]
    y_upper  = y_lower + size
    ax.fill_betweenx(np.arange(y_lower, y_upper),
                      0, ith_sil,
                      facecolor=palette[i], alpha=0.7,
                      label=f"Cluster {i}")
    y_lower  = y_upper + 10

avg_sil = sil_vals_per_sample.mean()
ax.axvline(avg_sil, color="red", ls="--", lw=1.5,
            label=f"Mean = {avg_sil:.3f}")
ax.set_xlabel("Silhouette Coefficient")
ax.set_ylabel("Cluster")
ax.set_title("Silhouette Plot — K-Means (k=5) on NSL-KDD")
ax.legend(fontsize=9)
plt.tight_layout()
plt.savefig("images/kmeans_silhouette_plot.png", dpi=150, bbox_inches="tight")
plt.show()
print(f"Mean silhouette score: {avg_sil:.4f}")
```

```python
# ─────────────────────────────────────────────────────────────────────────────
# Cluster profile — what does each cluster look like?
# ─────────────────────────────────────────────────────────────────────────────

df_eval = df[["category"]].copy()
df_eval["cluster"] = cluster_labels

print("Cluster composition (% of each true category in each cluster):")
profile = pd.crosstab(
    df_eval["cluster"],
    df_eval["category"],
    normalize="index"
).round(3) * 100

print(profile.to_string())

# Heatmap
fig, ax = plt.subplots(figsize=(9, 5))
sns.heatmap(profile, annot=True, fmt=".1f", cmap="YlOrRd",
            ax=ax, cbar_kws={"label": "% of cluster"})
ax.set_title("Cluster Composition — % of True Category per Cluster")
ax.set_xlabel("True Attack Category")
ax.set_ylabel("K-Means Cluster")
plt.tight_layout()
plt.savefig("images/kmeans_cluster_profile.png", dpi=150, bbox_inches="tight")
plt.show()

print("\nInterpretation: High % in a single column means the cluster maps cleanly")
print("to one attack category — indicating K-Means has discovered meaningful structure.")
```

## Section 6 — Work Log

### Entry 1 — Supervised vs Unsupervised Comparison

**Design decision:** Used the same NSL-KDD dataset as the KNN project deliberately.

**Reasoning:** This creates a direct comparison between supervised (KNN) and unsupervised (K-Means) approaches on identical data. The contrast is educationally valuable: KNN uses labels to learn boundaries; K-Means discovers structure with no label information at all.

**Impact:** The Adjusted Rand Index (ARI) measures how well K-Means clusters align with true labels — an ARI of 1.0 would mean perfect alignment, 0.0 means random. This gives an objective measure of how much structure K-Means can recover without supervision.

---

### Entry 2 — StandardScaler vs MinMaxScaler

**Decision:** Used StandardScaler (zero mean, unit variance) rather than MinMaxScaler.

**Reasoning:** K-Means uses Euclidean distance, which is sensitive to outliers. The NSL-KDD dataset has extreme outlier values in src_bytes and dst_bytes (some attacks send massive amounts of data). StandardScaler is more robust to outliers than MinMaxScaler because it doesn't force all values into [0,1] — extreme values are normalised by the standard deviation rather than being compressed against a fixed range.

**Impact:** Cluster separation improved with StandardScaler compared to MinMaxScaler on this dataset.

---

### Entry 3 — K-Means++ Initialisation

**Adjustment:** Used `init='k-means++'` with `n_init=20` (20 random restarts).

**Why:** The lecture notes explicitly state that K-Means is sensitive to initialisation: *'Clusters are sensitive to initial assignment of centroids'* and *'k-means is not a deterministic algorithm.'* K-Means++ spreads initial centroids far apart, reducing the chance of poor local minima. Running 20 restarts and keeping the best result adds robustness.

**Impact:** WCSS and silhouette scores were more consistent and higher with k-means++ vs random initialisation.

---

### Entry 4 — Choosing k

**Finding:** The elbow method did not produce a sharp elbow — a common challenge noted in lectures: *'Sometimes the k is not obvious using the elbow method.'*

**Resolution:** Used the Silhouette score as a secondary criterion. Also set k=5 to match the known number of attack categories, providing a meaningful comparison against ground truth. In a real-world scenario without labels, the silhouette score would be the primary guide.

---

### Entry 5 — Cluster Profile Analysis

**Addition:** Added a cluster composition heatmap showing what percentage of each true category falls into each discovered cluster.

**Finding:** Clusters with high concentration of a single attack type indicate that K-Means successfully discovered meaningful structure. DoS attacks, which have very distinctive traffic patterns (high packet rates, low bytes per connection), tend to cluster cleanly. R2L and U2R attacks are harder to separate as they appear more similar to normal traffic in the feature space.

**Analysis:** This directly demonstrates the value and limitations of unsupervised learning — K-Means can discover obvious structure but struggles with subtle attack patterns that require labelled examples to learn.
